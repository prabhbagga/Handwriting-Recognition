```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Reading and cleaning data

```python
dfx = pd.read_csv('./xdata.csv')
dfy = pd.read_csv('./ydata.csv')
x = dfx.values 
y = dfy.values 
x = x[:,1:] 
y = y[:,1:].reshape((-1,))
query_point = np.array([2,3])
plt.scatter(x[:,0],x[:,1],c=y)
plt.scatter(query_point[0],query_point[1],color='red')
plt.show()
```

# KNN Algo

```python
def distance(x1,x2):
    return np.sqrt((sum(x1-x2)**2))
```

```python
def KNN(x,y,query_point,k=5):
    size = x.shape[0]
    distance_list = []
    for i in range(size):
        distance_list.append((distance(x[i],query_point),y[i]))
    distance_list = sorted(distance_list)
    k_nearest_neighbours = distance_list[:k]
    k_nearest_neighbours = np.array(k_nearest_neighbours)
    vals = np.unique(k_nearest_neighbours[:,1],return_counts=True)
    index = vals[1].argmax()
    pred = vals[0][index]
    return pred
```

```python
KNN(x,y,[1,1])
```
